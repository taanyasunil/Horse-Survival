```python
# Dividing into test and training set:
from sklearn.model_selection import train_test_split
trainData, testData = train_test_split(horseData, test_size = 0.2, random_state = 42)
```

```python
# Adding a new column that shows the index which will make it easier to work with the dataset
trainData['index'] = range(1, len(trainData) + 1)
# Having a look at the training set:
trainData
```

```python
# Seeing how many missing values are there in the training dataset:
trainData.isnull().sum()
```

```python
# Dividing the training data into its data and target variables:
data = trainData.drop('outcome', axis = 1)
target = trainData['outcome']
target2 = trainData[['outcome', 'index']]
```

```python
# Can implement feature selection here:
```

```python
# On closer observation of the data we can see that: the hospital_number data is not important in predicting if the horse will
# survive or not, so dropping it from our data.
data = data.drop('hospital_number', axis = 1)
testData = testData.drop('hospital_number', axis = 1)
```

```python
# Filling up the missing values:
# Starting off with the numerical values, so removing the categorical data
data_num = data.drop(['surgery','age','temp_of_extremities','peripheral_pulse','mucous_membrane','capillary_refill_time','pain','peristalsis','abdominal_distention','nasogastric_tube','nasogastric_reflux','rectal_exam_feces','abdomen','abdomo_appearance','surgical_lesion','cp_data','index'], axis = 1)
data_num
```

```python
# Replacing these values with the mean:
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'mean')
imputer.fit(data_num)
```

```python
X = imputer.transform(data_num)
```

```python
data_num = pd.DataFrame(X, columns = data_num.columns)
```

```python
# Implementing Min-Max Scaling (Normalisation)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
F = scaler.fit_transform(data_num)
data_num = pd.DataFrame(F, columns = data_num.columns)
data_num['index'] = range(1, len(data_num) + 1)
data_num
```

```python
# Merging the data_num, data_cat and target based on the index column and then removing the index as we won't need it anymore.
trainData1 = pd.merge(data_num, data_cat, on='index')
trainData1 = pd.merge(trainData1, target2)
trainData1 = trainData1.drop('index', axis =1)
trainData1
```

```python
# Trying to predict the abdomo_appearance:
# Creating a training set with only the variables that have no null values and the
trainset = trainData1.drop(['surgery','age','temp_of_extremities','peripheral_pulse','mucous_membrane','capillary_refill_time','pain','peristalsis','abdominal_distention','nasogastric_tube','nasogastric_reflux','rectal_exam_feces','abdomen','surgical_lesion','cp_data'], axis = 1)
trainset_ = trainset.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

# Implementing One hot Encoding using Column Transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Creating the testset:
testset = trainset[trainset.isna().any(axis=1)]
testsetdata = testset.drop('abdomo_appearance', axis = 1)
# Doing One Hot Encoding:
ct2 = ColumnTransformer([('one-hot-encoder', OneHotEncoder(drop = 'first'), ['outcome'])], remainder = 'passthrough')
testsetdata = ct2.fit_transform(testsetdata)

# Creating the trainingset:
trainsettarget = trainset_[['abdomo_appearance']]
ohe = OneHotEncoder(drop = 'first')
trainsettarget = ohe.fit_transform(trainsettarget.abdomo_appearance.values.reshape(-1, 1)).toarray()
trainsetdata = trainset_.drop('abdomo_appearance', axis = 1)
ct = ColumnTransformer([('one-hot-encoder', OneHotEncoder(drop = 'first'), ['outcome'])], remainder = 'passthrough')
trainsetdata = ct.fit_transform(trainsetdata)
trainsetdata
```

```python
trainsettarget
```

From this, we can now predict the value of the missing categories of abdomo_appearance:

```python
# Implementing the KNearestNeighbors:-
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors = 3, p=1)
knn_classifier.fit(trainsetdata, trainsettarget)
knn_predictions = knn_classifier.predict(trainsetdata)
```

```python
# Checking the accuracy of this: 
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(trainsettarget, knn_predictions, normalize = True)
accuracy
```

```python
# Implementing Decision Trees to check performance:
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(trainsetdata, trainsettarget)
predict1 = tree.predict(trainsetdata)
# Checking the accuracy:
from sklearn.metrics import accuracy_score
accuracy1 = accuracy_score(trainsettarget, predict1)
print("The accuracy of the Decision Tree is : " +  str(accuracy1))
print("This is most likely overfitting.")


# Trying out the Extra Tree Classifier:
from sklearn.tree import ExtraTreeClassifier
tree2 = ExtraTreeClassifier()
tree2.fit(trainsetdata, trainsettarget)
predict2 = tree2.predict(trainsetdata)
# Checking the accuracy of this model:
from sklearn.metrics import accuracy_score
accuracy2 = accuracy_score(trainsettarget, predict2)
print("The accuracy of the Extra Tree Classifier is : " +  str(accuracy2))
print("This is most likely overfitting.")


# Trying out RadiusNeighborsClassifier from Neighbors
from sklearn.neighbors import RadiusNeighborsClassifier
rnc = RadiusNeighborsClassifier(p = 1)
rnc.fit(trainsetdata, trainsettarget)
predict4 = rnc.predict(trainsetdata)
# Checking the accuracy:
from sklearn.metrics import accuracy_score
accuracy4 = accuracy_score(trainsettarget, predict4)
print("The accuracy of the RadiusNeighborsClassifier is : " +  str(accuracy4))
print("It does not perform very well. The KNN Classifer had a better performance.")


# Trying out ensembles RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
tree3 = RandomForestClassifier()
tree3.fit(trainsetdata, trainsettarget)
predict5 = tree3.predict(trainsetdata)
# Checking the accuracy:
from sklearn.metrics import accuracy_score
accuracy5 = accuracy_score(trainsettarget, predict5)
print("The accuracy of the RandomForestClassifier is : " +  str(accuracy5))
print("It is overfitting.")


# Trying out ensembles RidgeClassifier
from sklearn.linear_model import RidgeClassifier
rc = RidgeClassifier(alpha = 0.01)
rc.fit(trainsetdata, trainsettarget)
predict6 = rc.predict(trainsetdata)
# Checking the accuracy:
from sklearn.metrics import accuracy_score
accuracy6 = accuracy_score(trainsettarget, predict6)
print("The accuracy of the RidgeClassifier is : " +  str(accuracy6))
print("It does not perform well.")
```
